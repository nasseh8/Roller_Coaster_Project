```python
 rc['Location'].value_counts()
```

```python
rc.query('Location!="Other"')
```

```python
rc.query('Location!="Other"').groupby('Location')['speed_mph']\
.agg(['mean','count']).query('count>=10')\
.sort_values('mean')
```

```python
ax=rc.query('Location!="Other"').groupby('Location')['speed_mph']\
.agg(['mean','count']).query('count>=10')\
.sort_values('mean')['mean'].plot(kind='barh', figsize=(12,5),\
title='Average Roller Coaster Speed By Location')
ax.set_xlabel('Average Roller Coaster Speed')
plt.show()
```

```python
rc['Manufacturer'].value_counts()
```

```python
rc['Manufacturer']
```

```python
rc.groupby('Manufacturer')['speed_mph']\
    .agg(['mean', 'count'])\
    .sort_values('mean', ascending=True)\
    .head(20)['mean'].plot(kind='barh', figsize=(12, 6), color='skyblue', 
                           title='Top 20 Manufacturers by Average Roller Coaster Speed')

# Customize the plot
plt.xlabel('Average Speed (mph)', fontsize=12)
plt.ylabel('Manufacturer', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
rc.groupby('Type_Main')['speed_mph']\
    .agg(['mean', 'count'])\
    .sort_values('mean', ascending=False)['mean'].plot(kind='bar', figsize=(12, 6), color='orange', 
                                                       title='Roller Coaster Material Types by Average Speed')
plt.xlabel('Average Speed (mph)', fontsize=12)
plt.ylabel('Material Type', fontsize=12)
plt.tight_layout()
plt.show()
```
